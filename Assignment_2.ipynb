```python
import pandas as pd

df = pd.read_csv('House Price India.csv')
df
```

```python
import matplotlib.pyplot as plt

# Histogram
plt.hist(df['number of bedrooms'], bins=10)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Bedrooms')
plt.show()
```

```python
plt.scatter(df['number of bedrooms'], df['Price'])
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Scatter Plot: Number of Bedrooms vs. Price')
plt.show()
```

```python
import seaborn as sns

selected_vars = ['Price', 'number of bedrooms', 'number of bathrooms', 'living area']
sns.pairplot(df[selected_vars])
plt.show()
```
